### 1 )

````python
class ABR:
    '''Un ABR référence un noeud ou rien. On suppose qu'un ABR ne peut
       contenir qu'une seule occurrence de chaque valeur.'''
    def __init__(self, cle = None):
        self.racine = cle
        self.gauche = None
        self.droit = None

    def est_vide(self):
        return self.racine is None
    
    def inserer(self, cle):
        '''Insertion d'une clé dans l'arbre.'''
        if self.est_vide():
            self.racine = cle
        else:
            if cle < self.racine:
                if self.gauche is None:
                    self.gauche = ABR(cle)
                else:
                    self.gauche.inserer(cle)
            elif cle > self.racine:
                if self.droit is None:
                    self.droit = ABR(cle)
                else:
                    self.droit.inserer(cle)

    def afficher(self, nl=True):  #  le paramètre nl permet d'ajouter un saut de ligne à la fin de l'affichage de l'ABR
        if not self.est_vide():
            if self.gauche is not None:
                self.gauche.afficher(nl=False)
            print(self.racine, end= ' ')
            if self.droit is not None:
                self.droit.afficher(nl=False)
            if nl:
                print()

    def rechercher(self, cle):
        if self.est_vide():
            return False
        elif self.racine == cle:
            return True
        elif self.gauche is not None and cle < self.racine:
            return self.gauche.rechercher(cle)
        elif self.droit is not None and cle > self.racine:
            return self.droit.rechercher(cle)
        else:
            return False
````

````python
from random import randint

liste = [randint(0, 100) for _ in range(20)]  # 20 entiers entre 0 et 99
print(liste)

mon_abr = ABR(liste[0])
for index in range(1, 20):
    mon_abr.inserer(liste[index])
````

## Exercice 3

````python
class Noeud:
    """Un nœud d'un arbre binaire"""
    def __init__(self, g, e, d):
        self.gauche = g
        self.etiquette = e
        self.droit = d
````


### 1)

````python
def inserer(arbre: Noeud, etiquette: tuple)->Noeud:
    '''Insertion d'une clé dans l'arbre.'''
    if arbre is None:
        return Noeud(None, etiquette, None)
    elif etiquette[0] < arbre.etiquette[0]:
        return Noeud(inserer(arbre.gauche, etiquette), arbre.etiquette, arbre.droit)
    elif etiquette[0] > arbre.etiquette[0]:
        return Noeud(arbre.gauche, arbre.etiquette, inserer(arbre.droit, etiquette))
    else:
        return arbre
````
