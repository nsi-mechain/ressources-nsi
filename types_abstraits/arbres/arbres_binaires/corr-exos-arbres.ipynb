### Question 8 :

```
Fonction hauteur(arbre):
    Si arbre = nil
        retourne 0
    Sinon
        retourne 1 + max(Hauteur Arbre.gauche, Hauteur Arbre.droite)
```

### Question 9 :

````python
def hauteur(arbre):
    if arbre == None:
        return 0
    else:
        return 1 + max(hauteur(arbre.gauche), hauteur(arbre.droit))

    
print(hauteur(arbre))
````

## Exercice 4 : implémentation du parcours en largeur en Python

````python
def parcours_largeur(arbre):
    file = [arbre]
    while file:
        noeud = file.pop(0)
        print(noeud.donnee)
        if noeud.gauche is not None:
            file.append(noeud.gauche)
        if noeud.droit is not None:
            file.append(noeud.droit)
                    
                   
parcours_largeur(arbre)
````

## Exercice 6 : hauteur d'un arbre complet

Voici le début du tableau (réalisé exprimentalement en procédant dans l'ordre `h` puis `N` puis `n`) :

|  N  |  n  |  h  |
|----:|:---:|:---:|
| 1   |  1  |  0  |
| 3   |  2  |  1  |
| 7   |  3  |  2  |
| 15  |  4  |  3  |
| 31  |  5  |  4  |
| 63  |  6  |  5  |
| 127 |  7  |  6  |

- `h = n - 1`

- `N = 2**n - 1` donc ` n = lb(N+1)` (voir `lb`: [logarithme binaire](https://fr.wikipedia.org/wiki/Logarithme_binaire))

- `h = lb(N+1) - 1`
