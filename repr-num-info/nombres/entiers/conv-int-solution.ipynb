### 3. En *Python*, concevez une **fonction** permettant de convertir un entier en un une chaîne de caractères le représentant en binaire sur 1 octet en *code complément à deux*

````python
def int2cp2(i: int) -> str:
    if -128 <= i <= +127:

        if i >= 0:
            return f'{i:08b}'
        else:
            return f'{256+i:08b}'
    else:
        return 'error'


# test
for i in (10, -10, 0, 66, -1, 127, 126, 128, -128, -200):
    print(f'{i:4} -> {int2cp2(i)}')
````

### 4. Idem pour la **fonction** réciproque 

````python
def cp22int(b: str) -> int:
    if b[0] == '0': # positif
        return int(b, 2)
    else: # négatif
        return int(b, 2) - 256


# test
for b in ('00000000', '00000001', '01111111', '11111111', '10000000', '00001010', '11110110'):
    print(f'{b} -> {cp22int(b):4}')
````
